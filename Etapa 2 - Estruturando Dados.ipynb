***Etapa 2 - Estruturando Dados***

Importar Bibliotecas


```
from tqdm import tqdm
import pandas as pd
import glob
import os
```

Coletando o Path dos arquivos de Musicas.TXT

```
path_lyrics = glob.glob(f'/content/drive/My Drive/Trabalho_DSD/data/*.txt')
path_lyrics[1:10]
```

Ler Conteúdo dos Arquivos e Mapear em DataFrame

```
def file_to_dataframe(path_lyrics):
    
    collection_lyrics = []
    for path in tqdm(path_lyrics):
        #Ler conteúdo do arquivo .txt
        with open(path,'r') as f:
            
            #conteudo do arquivo txt, que é a letra da música completa
            file = f.read()
            print(path)
            
            obj = path.split(os.sep)[6].split("$")
            #Coleta nome da banda  pelo path
            band_name = obj[0]
            #Coleta letra da música pelo path
            lyric_name = obj[1].replace(".txt", "")
            
            collection_lyrics.append({
                "text" : file,
                "band" : band_name,
                "lyric" : lyric_name
            })
    return pd.DataFrame.from_dict(collection_lyrics)
```

```
df_lyrics = file_to_dataframe(path_lyrics)
```
