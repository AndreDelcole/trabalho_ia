Análise de Distribuiçao por Estilo Musical

```
    x1 = df_lyrics.loc[df_lyrics['band'] == "coral-kemuel", 'band']
    x2 = df_lyrics.loc[df_lyrics['band'] == "fernandinho", 'band']
    x3 = df_lyrics.loc[df_lyrics['band'] == "roupa-nova", 'band']
    x4 = df_lyrics.loc[df_lyrics['band'] == "legiao-urbana", 'band']
    x5 = df_lyrics.loc[df_lyrics['band'] == "guns-n-roses", 'band']
    x6 = df_lyrics.loc[df_lyrics['band'] == "aerosmith", 'band']

    trace1 = go.Histogram(x=x1, name='Gospel', opacity=0.75)
    trace2 = go.Histogram(x=x2, name='Gospel', opacity=0.75)
    trace3 = go.Histogram(x=x3, name='Rock Nacional', opacity=0.75)
    trace4 = go.Histogram(x=x4, name='Rock Nacional', opacity=0.75)
    trace5 = go.Histogram(x=x5, name='Rock Internacional', opacity=0.75)
    trace6 = go.Histogram(x=x6, name='Rock Internacional', opacity=0.75)

    data = [trace1, trace2, trace3, trace4, trace5, trace6]
    layout = go.Layout(barmode='overlay', title='Análise de Distribuição por Estilo Musical')
    fig = go.Figure(data=data, layout=layout)

    iplot(fig, filename='overlaid histogram')
```
